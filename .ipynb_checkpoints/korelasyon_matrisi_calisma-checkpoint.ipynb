```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("mushrooms.csv")
data.shape
```

```python
#X'ten hedefimiz olan class'ı ayırıyoruz, y hedef olduğu için sadece class kalıyor.
X = data.drop(["class"], axis=1)
y = data["class"]

#X'e one hot encoding yapıyoruz
X_encoded = pd.get_dummies(X, prefix_sep="_")
#y'ye label encoding yapıyoruz
y_encoded = LabelEncoder().fit_transform(y)
#encode edilmiş olarak class'ı X'e ekliyoruz;
X_encoded["class"] = y_encoded
```

```python
X_encoded.corr()
```

```python
import seaborn as sns
```

```python
sns.heatmap(X_encoded.iloc[:,-7:].corr(), annot=True)
```

```python
#korelasyonu en yüksek 15 kolon
X_encoded.corr()["class"].abs().nlargest(15)
```

```python
X_highest_cor = X_encoded.corr()["class"].abs().nlargest(10).index
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6), dpi=150)
sns.heatmap(X_encoded[X_highest_cor].corr().abs(), annot=True);
```
